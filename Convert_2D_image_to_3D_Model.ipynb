```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
image_path = '/content/laptop.jpeg'
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
scale_percent = 10
width = int(image.shape[1] * scale_percent / 100)
height = int(image.shape[0] * scale_percent / 100)
resized_image = cv2.resize(image, (width, height))
```

```python
fig = plt.figure()
ax = fig.add_subplot(121, projection='3d')

x = np.linspace(0, width - 1, width)
y = np.linspace(0, height - 1, height)
x, y = np.meshgrid(x, y)

z = resized_image[:, :, 0] / 255.0

ax.plot_surface(x, y, z, cmap='viridis')
ax.set_title("3D Model")

ax2 = fig.add_subplot(122)
ax2.imshow(image_rgb)
ax2.set_title("Laptop IN 2D")

plt.tight_layout()

image_height_pixels = height
image_width_pixels = width

object_width_pixels = 100
object_real_width = 15.0
pixel_to_real_ratio = object_real_width / object_width_pixels

image_real_height = image_height_pixels * pixel_to_real_ratio
image_real_width = image_width_pixels * pixel_to_real_ratio

plt.show()
```

```python
print(f"Image Real Dimensions: Height = {image_real_height:.2f} inches, Width = {image_real_width:.2f} inches")
```
